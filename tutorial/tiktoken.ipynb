```python
import tiktoken
```

```python
# Load a specific encoding by name
enc = tiktoken.get_encoding("gpt2")
```

```python
text = "Wait, don't do it!"

# ENCODE: Convert text to a list of integer IDs
token_ids = enc.encode(text)
# Output: [11339, 11, 836, 470, 466, 340, 0]

token_ids
```

```python
# DECODE: Convert a list of integer IDs back to a single string
decoded_text = enc.decode(token_ids)
# Output: "Wait, don't do it!"

decoded_text
```

```python
text_with_special = "End of sequence <|endoftext|>"

# Allow specific special tokens
ids = enc.encode(text_with_special, allowed_special={"<|endoftext|>"})

ids
```

```python
# Or allow all special tokens registered in that encoding
ids_all = enc.encode(text_with_special, allowed_special="all")

ids_all
```

```python
# 1. Vocabulary Size
# CRITICAL: Pass this exact integer to your nn.Embedding(vocab_size, embedding_dim)
vocab_size = enc.n_vocab 
# For 'gpt2', this is 50257

vocab_size
```

```python
# 2. Inspecting a Single Token
# Great for debugging to see exactly what characters/bytes a specific ID maps to
byte_string = enc.decode_single_token_bytes(11339)
# Output: b'Wait'

byte_string
```

```python
# 3. Special Token Dictionary
# See which special tokens are natively supported by the loaded encoding
special_tokens = enc._special_tokens

special_tokens
```
